```python
# 形態素解析
from janome.tokenizer import Tokenizer 
# wikipediaをインストール
import wikipedia

# 言語を日本語に設定
wikipedia.set_lang("JP")

# defはfunctionのようなもの、関数を作る
def search_meaning(keyword):
    # wikiで検索したものをtarget_keywordへ格納
    target_keyword = wikipedia.search(keyword)
    # target_keywordがwikiにない場合「見つかりません」と出力
    if not target_keyword:
        return"見つかりません"
    # wikiにtarget_keywordがある場合（ないことがない場合）「１（０）番目」に出てきた結果を出力する
    else :
        return wikipedia.summary(target_keyword[0])
    
tokenizer = Tokenizer()
    
def get_response(user_input):
    # ここでユーザーの入力に対する処理を行う
    # ユーザーが入力した情報を形態素解析
    tokens = tokenizer.tokenize(user_input)

# 調べて返す
    for token in tokens:
        if token.part_of_speech.split(",")[1] == "固有名詞":
            meaning = search_meaning(token.surface)
            return meaning
        return "何か知らない言葉はありますか？"
    

def search_meaning(term):
    try:
        return wikipedia.summary(term)
    except wikipedia.DisambiguationError as e:
        # 候補を最大5つ表示
        options = e.options[:5]
        print("複数の候補があります。どれを知りたいですか？")
        for i, option in enumerate(options, 1):
            print(f"{i}. {option}")
        
        # ユーザーに選択させる
        while True:
            try:
                choice = int(input("番号を入力してください (1-5): ")) - 1
                if 0 <= choice < len(options):
                    return wikipedia.summary(options[choice])
                else:
                    print("番号が範囲外です。1～5の間で選んでください。")
            except ValueError:
                print("数字を入力してください。")
    
# メインの対話ループの処理
print("ボット：名前を入力してください！")
user_name = input()
print (f"ボット：こんにちは！{user_name}さん")

print("ボット：私は単語回答ボットです！知らない単語を質問してください")
# while文のTrueで処理が終わるまで回り続けるようにする
while True:
    user_input = input("ユーザー：  *終了するには「終了」と。")
    if user_input == "終了":
        print(f"ボット：チャットを終了します。")
        break
    
    response = get_response(user_input)
    print(f"ボット：{response}")
```
